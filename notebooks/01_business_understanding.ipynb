# 1. What is Customer Churn?

Customer churn is the process in which an existing customer stops using a company's product or service.

A customer is considered "churned" when they discontinue their relationship with the business, such as cancelling a subscription, switching to a competitor, or closing their account.

Examples:
- A Netflix user cancels their subscription.
- A Jio customer switches to Airtel.
- A bank customer closes their savings account.
- A customer stops purchasing from an online store.

For businesses, customer churn represents lost revenue and reduced long-term profitability. Therefore, understanding and predicting churn is a critical business objective.


# 2. Why is Customer Churn Important?

Customer retention is generally much less expensive than acquiring new customers.

For example, suppose a telecom customer pays ₹1,000 per month.

If the customer stays for three years:

Revenue = ₹1,000 × 36 = ₹36,000

If that customer leaves after one year, the company loses the remaining future revenue.

By identifying customers who are likely to churn, businesses can take proactive actions such as:
- Offering discounts
- Providing better customer support
- Recommending suitable plans
- Improving customer experience

Reducing churn directly increases revenue and customer lifetime value.

# 3. Types of Customer Churn

Customer churn can be broadly classified into two categories.

### Voluntary Churn

The customer decides to leave the company.

Reasons include:
- Better competitor offers
- High prices
- Poor customer service
- Poor product quality

### Involuntary Churn

The company terminates the customer relationship.

Examples include:
- Fraudulent activities
- Non-payment
- Policy violations

In this project, our primary focus is predicting voluntary churn because businesses can often prevent it.
