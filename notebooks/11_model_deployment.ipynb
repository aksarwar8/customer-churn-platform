# Customer Churn Model Deployment

## Objective

The objective of this notebook is to demonstrate how the trained CatBoost model can be used to predict customer churn for new customer records.

The notebook covers:

- Loading the trained model
- Making predictions
- Predicting churn probability
- Batch inference
- Saving prediction results

This notebook simulates the inference stage of a production machine learning system.

```python
import joblib
import pandas as pd
```

```python
model = joblib.load("../models/catboost_pipeline.pkl")

print("Model loaded successfully!")
```

```python
df = pd.read_csv("../data/processed/telco_feature_engineered.csv")

X = df.drop(
    columns=[
        "customerID",
        "Churn",
        "Churn_Binary"
    ]
)

y = df["Churn_Binary"]
```

```python
customer = X.iloc[[0]]

prediction = model.predict(customer)

probability = model.predict_proba(customer)

print("Prediction :", prediction)

print("Probability :", probability)
```

```python
sample = X.iloc[:10]

predictions = model.predict(sample)

probabilities = model.predict_proba(sample)

results = sample.copy()

results["Prediction"] = predictions

results["Stay Probability"] = probabilities[:,0]

results["Churn Probability"] = probabilities[:,1]

results.head(10)
```

```python
pred = model.predict(X)

prob = model.predict_proba(X)

deployment_results = X.copy()

deployment_results["Prediction"] = pred

deployment_results["Stay Probability"] = prob[:,0]

deployment_results["Churn Probability"] = prob[:,1]

deployment_results.head()
```

```python
deployment_results.to_csv(
    "../reports/churn_predictions.csv",
    index=False
)

print("Predictions saved successfully!")
```
